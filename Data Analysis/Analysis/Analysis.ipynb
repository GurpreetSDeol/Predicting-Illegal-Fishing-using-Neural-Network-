```python
import geopandas as gpd
import matplotlib.pyplot as plt 
import pandas as pd 
import matplotlib.patches as mpatches

```

Load all pf the data files

```python
ship_data_path =rf'D:\Datasets\Illegal Fishing\Processed Data\filtered_ship_data.csv'
ship_data = pd.read_csv(ship_data_path)


gdb_path_land = rf'D:\Datasets\Illegal Fishing\Original Data\ne_10m_land\ne_10m_land.shp'
land_data = gpd.read_file(gdb_path_land)

gdb_path_MPZ = rf'D:\Datasets\Illegal Fishing\Original Data\WDPA_Jun2024_Public.gdb'
layers =  ['WDPA_poly_Jun2024', 'WDPA_point_Jun2024', 'WDPA_source_Jun2024']


poly_within_ocean = rf'D:\Datasets\Illegal Fishing\Processed Data\Filtered MPA data\filtered_polygons.shp'
poly_data = gpd.read_file(poly_within_ocean)

points_within_ocean = rf'D:\Datasets\Illegal Fishing\Processed Data\Filtered MPA data\filtered_points.shp'
point_data = gpd.read_file(points_within_ocean)

```

Plot Latitude and Longitude data alongside Marine Protected Zones

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_xlim(-180, 180)
ax.set_ylim(-90, 90)
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')

land_data.plot(ax=ax, color='lightgrey', edgecolor='black')
point_data.plot(ax=ax, color = 'white', edgecolor = 'green')
poly_data.plot(ax=ax, color = 'white', edgecolor = 'green')

ax.set_title('Oceanic Marine Protected Zones')
fig.savefig(rf'C:\Users\gurpr\OneDrive\Documents\New Projects\Illegal Fishing\Data Analysis\Plots\Oceanic Marine Protected Zones.png')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_xlim(-180, 180)
ax.set_ylim(-90, 90)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')


land_data.plot(ax=ax, color='lightgrey', edgecolor='black')

point_data.plot(ax=ax, color = 'white', edgecolor = 'green')
poly_data.plot(ax=ax, color = 'white', edgecolor = 'green')

ship_data_yes = ship_data[ship_data['illegal']=='yes']
ship_data_no = ship_data[ship_data['illegal']=='no']

ax.scatter(ship_data_no['lon'],ship_data_no['lat'],s=0.03,c='blue', label = 'Legal')
ax.scatter(ship_data_yes['lon'],ship_data_yes['lat'],s=0.03,c='red',label = 'Illegal')


red_patch = mpatches.Patch(color='red', label='Illegal Fishing')
blue_patch = mpatches.Patch(color='blue', label='Legal Fishing')

ax.legend(handles=[red_patch, blue_patch], loc='lower left')

ax.set_title(' Fishing events and Oceanic Marine Protected Zones in 2020')
fig.savefig(rf'C:\Users\gurpr\OneDrive\Documents\New Projects\Illegal Fishing\Data Analysis\Plots\Fishing events and Oceanic Marine Protected Zones in 2020.png')

```
